## Let's binarize our labels

```python
data["Tags"].iloc[0:10]
```

```python
target_interest  = ["<c#>", "<c++>", "<java>", "<javascript>", "<html>","<css>","<android>",
                    "<python>", "<r>", "<sql>"]
```

```python
datatest = data
```

```python
#4min
clean_tags = []
tmp_tags = []
for i in range(data["Tags"].shape[0]):
  raw_content = data["Tags"].iloc[i]
  print(raw_content)
  for tag in target_interest:
    
    if tag in raw_content:
      #print(tag, tag in raw_content)
      tmp_tags.append(tag)
      print(tmp_tags)
  clean_tags.append(tmp_tags)
  tmp_tags = []
print(clean_tags)
```

```python
datatest["clean_tags"] = clean_tags
datatest
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
#save this into drive
target = mlb.fit_transform(clean_tags)


print(len(target_interest),target.shape,mlb.classes_)
target_df = pd.DataFrame(target,
                   columns=mlb.classes_)
target_df
```
